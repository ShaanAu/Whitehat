### My First<sup>TM</sup> Clustering Algorithm

1. Pick a value for k (the number of clusters to create)
2. Initialize k 'centroids' (starting points) in your data
3. Create your clusters. Assign each point to the nearest centroid. 
4. Make your clusters better. Move each centroid to the center of its cluster. 
5. Repeat steps 3-4 until your centroids converge.

# Let's practice a toy example by hand. Take a 1-dimensional array of numbers:

```python
arr = [2, 5, 6, 8, 12, 15, 18, 28, 30]
```

```python
import numpy as np
arr = np.array(arr)

cluster_1_centroid = np.mean(arr[:4])
cluster_2_centroid = np.mean(arr[4:7])
cluster_3_centroid = np.mean(arr[7:])
print(f'cluster 1 Centroid: {cluster_1_centroid}')
print(f'cluster 2 Centroid: {cluster_2_centroid}')
print(f'cluster 3 Centroid: {cluster_3_centroid}')
```

```python
cluster_1_inert = np.square(arr[:4] - cluster_1_centroid )
cluster_2_inert = np.square(arr[4:7] - cluster_2_centroid )
cluster_3_inert = np.square(arr[7:] - cluster_3_centroid )

inertia = np.sum(cluster_1_inert) + np.sum(cluster_2_inert) + np.sum(cluster_3_inert)
inertia
```

```python
cluster_1_inert
```

Pick k=3 random starting centroids and let the other points fall into clusters based on the closest centroid. Then, reassign your centroids based on the mean value of your cluster and repeat the process.

#### K Means is a powerful algorithm, but different starting points may give you different clusters. You won't necessarily get an optimal cluster.  

# Metrics for assessing your clusters

**Inertia** -- sum of squared errors for each cluster
- ranges from 0 to very high values
- low inertia = dense clusters
- lower is better!

**Silhouette Score** -- measure of how far apart clusters are
- ranges from -1 to 1
- high silhouette Score = clusters are well separated


**Inertia** -- sum of squared errors for each cluster
- low inertia = dense cluster

$$\sum_{j=0}^{n} (x_j - \mu_i)^2$$

where $\mu_i$ is a cluster centroid. (K-means explicitly tries to minimize this.)

`.inertia_` is an attribute of sklearn's kmeans models.


**Silhouette Score** -- measure of how far apart clusters are
- ranges from -1 to 1
- high silhouette Score = clusters are well separated

The definition is a little involved$^*$, but intuitively the score is based on how much closer data points are to their own clusters than to the nearest neighbor cluster.

We can calculate it in sklearn with `metrics.silhouette_score(X_scaled, labels, metric='euclidean')`.

$^*$<https://en.wikipedia.org/wiki/Silhouette_(clustering)>

# Guided practice
Let's do some clustering with the iris dataset.

```python
# Check out the dataset and our target values
df = pd.read_csv("./data/iris.csv")
df
```

```python
df.Name.value_counts()
```

Let's plot the data to see the distributions:

```python
cols = df.columns[:-1]
sns.pairplot(df[cols])
```

Next, since each of our features have different units and ranges, let's do some preprocessing:

We have seen **`standardscaler()`**. Another preprocessing method is **`MinMaxScaler()`** which can also be used to scale the data. You can find the two scalers in the **`preprocessing`** class of **`sklearn`**

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
# initialize the scaler

sc = StandardScaler()


# fit the scaler to data
sc.fit(df[cols])

# transform data
X_scaled = sc.transform(df[cols])

```

```python
X_scaled
```

```python
pd.DataFrame(X_scaled, columns=cols).describe()
```

Now that we've formatted our data and understand its structures, we can finally go ahead and cluster.

We're going to set k = 2, given the pattern we were seeing above in our graphs.

```python
# import KMeans class from sklearn
from sklearn.cluster import KMeans

# specify the number of k
k = 3

# Initialize model
K_means = KMeans(n_clusters=k, random_state=42)

# fit the data
K_means.fit(X_scaled)
```

We can use Scikit's built-in functions to determine the locations of the labels, centroids, and cluster inertia:

```python
# label_ return the labels of the cluster
labels = K_means.labels_

# cluster_centers_ returns the center pint for each cluster
centroids = K_means.cluster_centers_

# inertia measures the density of the cluster. remember lower is better!
inertia = K_means.inertia_


inertia, centroids, labels
```

And to compute the clusters' silhouette coefficient:


```python
from sklearn.metrics import silhouette_score

silhouette_score(X_scaled, labels, metric='euclidean')
```

...and we're done! You've completed your first clustering analysis.

Let's see how it looks. First, let's put the labels columns into our dataframe

```python
# create a new columns with the labels
df['labels'] = labels

```

```python
df.labels.value_counts()
```

```python
cols = df.columns[:-2]
```

Let's plot each cluster in a different color. Seaborn has a 'hue' parameter we can use for this. 

```python
sns.distributions._has_statsmodels = False
```

```python
sns.pairplot(data=df,x_vars=cols, y_vars=cols, hue='labels')
```

For comparison, here's the data colored by name of the plant. 


```python
sns.pairplot(data=df,x_vars=cols, y_vars=cols, hue='Name')
```

## How can you determine the optimum number of clusters?

We use a method known as the **elbow method**

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Independent practice

Repeat our clustering analysis for the foods nutrients dataset (below). There are no "true" labels for this one!

```python
df_nut = pd.read_csv('./data/nutrients.txt', sep=r'\s+')
df_nut
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
# initialize the scaler

sc_nut = StandardScaler()

cols = df_nut.columns[1:]

# fit the scaler to data
sc_nut.fit(df_nut[cols])

# transform data
X_scaled_nut = sc_nut.transform(df_nut[cols])
```

```python
wcss_nut = []
for i in range(1, 11):
    kmeans_nut = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans_nut.fit(X_scaled_nut)
    wcss_nut.append(kmeans_nut.inertia_)
plt.plot(range(1, 11), wcss_nut)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# import KMeans class from sklearn
from sklearn.cluster import KMeans

# specify the number of k
k_nut = 4

# Initialize model
K_means_nut = KMeans(n_clusters=k_nut, random_state=42)

# fit the data
K_means_nut.fit(X_scaled_nut)
```

```python
# label_ return the labels of the cluster
labels_nut = K_means_nut.labels_

# cluster_centers_ returns the center pint for each cluster
centroids_nut = K_means_nut.cluster_centers_

# inertia measures the density of the cluster. remember lower is better!
inertia_nut = K_means_nut.inertia_


inertia_nut, centroids_nut, labels_nut
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_scaled_nut, labels_nut, metric='euclidean')
```
